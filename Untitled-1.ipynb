```python
import requests
import pandas as pd

def fetch_const_disaster_df(service_key: str,
                           dsstr_dy: str,
                           call_api_id: str = "1010",
                           page_no: int = 1,
                           num_of_rows: int = 100,
                           timeout: int = 30) -> pd.DataFrame:
    """
    건설업 일별 중대재해 현황 API 호출 -> DataFrame 반환
    dsstr_dy: 'YYYYMMDD' 문자열
    """
    url = "https://apis.data.go.kr/B552468/constDsstr01/getconstDsstr01"
    params = {
        "serviceKey": service_key.strip(),
        "dsstrDy": dsstr_dy,
        "callApiId": call_api_id,
        "pageNo": str(page_no),
        "numOfRows": str(num_of_rows),
        "_type": "json",
    }

    r = requests.get(url, params=params, timeout=timeout)

    # 응답이 이상할 때 원인 파악 쉽게
    if r.status_code != 200:
        raise RuntimeError(f"HTTP {r.status_code}\n{r.text[:500]}")

    # JSON 파싱
    data_json = r.json()

    # item 꺼내기(빈 결과/단일 dict/리스트 모두 대응)
    body = data_json.get("body", {})
    items = body.get("items", {}).get("item", [])
    if isinstance(items, dict):
        items = [items]
    if items is None:
        items = []

    df = pd.DataFrame(items)

    # 자주 쓰는 타입 변환(있을 때만)
    if "dsstrDy" in df.columns:
        df["dsstrDy"] = pd.to_datetime(df["dsstrDy"].astype(str), format="%Y%m%d", errors="coerce")

    return df


# ===== 사용 예시 =====
service_key = "BwUR8QqbORa9AHaaFBq84E9jlctfN6DCXzXyEFt179E44ErIEysOU94AXtsamsvhHKtWt1ZHsJa/h2ZBaZyXfw=="     # 디코딩키 권장
dsstr_dy = "20190827"                 # 조회할 날짜(YYYYMMDD)

df = fetch_const_disaster_df(service_key, dsstr_dy, num_of_rows=100)
data = df  # 네가 말한 data 함수/변수에 넣기

print(df.shape)
print(df.columns.tolist())
display(df.head(5))
```

```python
import time
import requests
import pandas as pd
from datetime import datetime, timedelta

BASE_URL = "https://apis.data.go.kr/B552468/constDsstr01/getconstDsstr01"

def fetch_page(service_key: str, dsstr_dy: str, page_no: int, num_of_rows: int, call_api_id: str = "1010"):
    params = {
        "serviceKey": service_key.strip(),
        "dsstrDy": dsstr_dy,          # YYYYMMDD
        "callApiId": call_api_id,
        "pageNo": str(page_no),
        "numOfRows": str(num_of_rows),
        "_type": "json"
    }
    r = requests.get(BASE_URL, params=params, timeout=30)
    r.raise_for_status()
    j = r.json()

    body = j.get("body", {})
    total = int(body.get("totalCount", 0) or 0)

    items = body.get("items", {}).get("item", [])
    if isinstance(items, dict):
        items = [items]
    if items is None:
        items = []

    return total, items

def fetch_all_for_date(service_key: str, dsstr_dy: str, num_of_rows: int = 1000,
                       sleep_sec: float = 0.15, max_retries: int = 3) -> pd.DataFrame:
    # 1) 1페이지로 totalCount 확인
    for attempt in range(1, max_retries + 1):
        try:
            total, items = fetch_page(service_key, dsstr_dy, page_no=1, num_of_rows=num_of_rows)
            break
        except Exception:
            if attempt == max_retries:
                raise
            time.sleep(0.5 * attempt)

    if total == 0:
        return pd.DataFrame()

    all_items = list(items)

    # 2) 나머지 페이지 수집
    pages = (total + num_of_rows - 1) // num_of_rows
    for p in range(2, pages + 1):
        time.sleep(sleep_sec)
        for attempt in range(1, max_retries + 1):
            try:
                _, more = fetch_page(service_key, dsstr_dy, page_no=p, num_of_rows=num_of_rows)
                all_items.extend(more)
                break
            except Exception:
                if attempt == max_retries:
                    raise
                time.sleep(0.5 * attempt)

    df = pd.DataFrame(all_items)

    # 날짜 컬럼 정리
    if "dsstrDy" in df.columns:
        df["dsstrDy"] = pd.to_datetime(df["dsstrDy"].astype(str), format="%Y%m%d", errors="coerce")

    return df

def daterange(start_yyyymmdd: str, end_yyyymmdd: str):
    s = datetime.strptime(start_yyyymmdd, "%Y%m%d").date()
    e = datetime.strptime(end_yyyymmdd, "%Y%m%d").date()
    cur = s
    while cur <= e:
        yield cur.strftime("%Y%m%d")
        cur += timedelta(days=1)

# =========================
# 실행: 2021-01-01 ~ 2021-12-31 전수 수집 -> data에 저장
# =========================
service_key = "BwUR8QqbORa9AHaaFBq84E9jlctfN6DCXzXyEFt179E44ErIEysOU94AXtsamsvhHKtWt1ZHsJa/h2ZBaZyXfw=="  # 디코딩키 권장

start_date = "20210101"
end_date   = "20211231"

dfs = []
for d in daterange(start_date, end_date):
    df_d = fetch_all_for_date(service_key, d, num_of_rows=1000, sleep_sec=0.15)
    if not df_d.empty:
        dfs.append(df_d)
    print(f"{d} done. rows={len(df_d)}")

data = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

print("ALL DONE:", data.shape)
display(data.head())
```

```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET

def response_to_df(r: requests.Response) -> pd.DataFrame:
    ct = (r.headers.get("Content-Type") or "").lower()
    text = r.text.strip()

    # JSON일 때
    if "json" in ct or (text.startswith("{") and text.endswith("}")):
        j = r.json()
        # (기상청/공공데이터포털 공통) response->body->items->item 패턴이 많음
        items = (
            j.get("response", {})
             .get("body", {})
             .get("items", {})
             .get("item", [])
        )
        if isinstance(items, dict):
            items = [items]
        return pd.DataFrame(items)

    # XML일 때
    if text.startswith("<?xml") or "<response" in text[:200]:
        root = ET.fromstring(text)
        rows = []
        for item in root.iter("item"):
            rows.append({child.tag: child.text for child in item})
        return pd.DataFrame(rows)

    # 그 외(에러/빈 응답)
    raise RuntimeError(f"Not JSON/XML. status={r.status_code}\n{text[:500]}")

# 사용 예시
r = requests.get(url, params=params, timeout=30)
print("status:", r.status_code, "ct:", r.headers.get("Content-Type"))
df_weather = response_to_df(r)
df_weather.head()
```

```python
import re
import pandas as pd
import numpy as np

def parse_kma_start7777_to_df(text: str) -> pd.DataFrame:
    lines = text.splitlines()

    # 1) 헤더에서 컬럼 약어 추출: "# 12. TA : 기온 (C)" 같은 줄
    col_pattern = re.compile(r"^#\s*\d+\.\s*([A-Z0-9_]+)\s*:")
    cols = []
    for line in lines:
        m = col_pattern.match(line.strip())
        if m:
            cols.append(m.group(1))

    # 컬럼 추출이 안되면 중단
    if not cols:
        raise ValueError("컬럼 약어를 헤더(# ...)에서 찾지 못했습니다. 응답 포맷이 다를 수 있어요.")

    # 2) 데이터 라인: '#'로 시작하지 않는 줄만 모음
    data_lines = []
    for line in lines:
        s = line.strip()
        if not s:
            continue
        if s.startswith("#"):
            continue
        # 가끔 END/START 같은 제어문이 있으면 제외
        if "END" in s.upper() and len(s) < 30:
            continue
        data_lines.append(s)

    if not data_lines:
        # 데이터가 정말 없는 경우
        return pd.DataFrame(columns=cols)

    # 3) 공백(연속 공백 포함) 기준 split → rows 만들기
    rows = []
    for s in data_lines:
        parts = re.split(r"\s+", s)
        # 컬럼 수와 값 수가 다르면 (뒤에 설명이 붙거나) 앞쪽만 취함
        if len(parts) >= len(cols):
            parts = parts[:len(cols)]
        else:
            # 부족하면 NaN으로 채움
            parts = parts + [None] * (len(cols) - len(parts))
        rows.append(parts)

    df = pd.DataFrame(rows, columns=cols)

    # 4) 타입 정리
    # TM(관측시각) → datetime (예: 201007151200)
    if "TM" in df.columns:
        df["TM"] = pd.to_datetime(df["TM"], format="%Y%m%d%H%M", errors="coerce")

    # 숫자형으로 바꿀 수 있는 컬럼은 최대한 numeric 변환
    for c in df.columns:
        if c == "TM":
            continue
        df[c] = pd.to_numeric(df[c], errors="coerce")

    return df


# ===== 사용 예시 =====
# r = requests.get(...)

df_weather = parse_kma_start7777_to_df(r.text)
print(df_weather.shape)
display(df_weather.head())
```
